## Helper function to calculate standard deviation from a list of percentages

```python
def calculate_stats(percentage_lists):
    """Calculate mean and standard deviation from a list of percentages."""
    stats = {}
    for key, percentages in percentage_lists.items():
        if percentages:
            stats[key] = {
                'Percentage': np.mean(percentages),
                'Std Dev': np.std(percentages)
            }
        else:
            stats[key] = {
                'Percentage': 0,
                'Std Dev': 0
            }
    return stats
```

## Helper function to calculate the covariance matrix between beta and gamma

Since they are both extracted from the same dataset there can be a correlation

```python
def calculate_covariance(dfs_clean):
    covariance_results = {}

    for socket in ['LGA', 'BGA']:
        # Combine all dataframes for the current socket
        socket_df = pd.concat([df[df['Socket'] == socket] for df in dfs_clean], ignore_index=True)
        if socket_df.empty:
            continue

        # Get unique high-level materials and substances
        high_level_materials = socket_df['High Level Material'].unique()
        substances = socket_df['Substances'].unique()

        # Initialize a dictionary to store the data for each high-level material and substance
        data_dict = {material: {sub: [] for sub in substances} for material in high_level_materials}

        # Populate the data dictionary with the percentages
        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    data_dict[material][sub].append(percentage)

        # Create a DataFrame for each sample (row in the original data)
        sample_data = []
        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            sample_data_row = {}
            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()
                sample_data_row[material] = total_mass / sample_socket_df['CPU Total Mass [mg]'].iloc[0] if sample_socket_df['CPU Total Mass [mg]'].iloc[0] else 0

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    sample_data_row[f"{material}_{sub}"] = percentage

            sample_data.append(sample_data_row)

        # Create a DataFrame from the sample data
        sample_df = pd.DataFrame(sample_data)

        # Calculate the covariance matrix
        cov_matrix = sample_df.cov().fillna(0)

        # Extract the covariance between beta (high-level materials) and gamma (substances)
        n_beta = len(high_level_materials)
        cov_beta = cov_matrix.loc[high_level_materials, high_level_materials].values
        cov_gamma = np.zeros((n_beta, n_beta))
        cov_beta_gamma = np.zeros((n_beta, n_beta))

        for i, material_i in enumerate(high_level_materials):
            for k, material_k in enumerate(high_level_materials):
                # Covariance between material_i and material_k's substances
                for sub in substances:
                    cov_gamma[i, k] += cov_matrix.loc[f"{material_i}_{sub}", f"{material_k}_{sub}"]
                    cov_beta_gamma[i, k] += cov_matrix.loc[material_i, f"{material_k}_{sub}"]

        # Normalize the covariance matrices
        cov_gamma /= len(substances)
        cov_beta_gamma /= len(substances)

        # Store the covariance matrices for the current socket
        covariance_results[socket] = {
            'cov_beta': cov_beta,
            'cov_gamma': cov_gamma,
            'cov_beta_gamma': cov_beta_gamma
        }

    return covariance_results
```

## **Calculation of $\vec\beta$**

$\vec\beta : (5 \times 1)$

$\Sigma_i \beta_i = 1$, i = 1,2,3,4,5

Linear regression: % of the total mass of the CPU for each high level component


```python
def high_level_mass_matrix(dfs_clean):
    results = {}

    for socket in ['LGA', 'BGA']:
        percentage_lists = {material: [] for material in pd.concat(dfs_clean)['High Level Material'].unique()}

        for df in dfs_clean:
            socket_df = df[df['Socket'] == socket]
            if socket_df.empty:
                continue

            total_mass = socket_df['Mass [mg]'].sum()
            for material in percentage_lists.keys():
                material_mass = socket_df[socket_df['High Level Material'] == material]['Mass [mg]'].sum()
                percentage = material_mass / total_mass if total_mass else 0
                percentage_lists[material].append(percentage)

        stats = calculate_stats(percentage_lists)
        results[socket] = pd.DataFrame.from_dict(stats, orient='index')

    return results
```

## **Calculation of matrix **$\gamma$****

$\gamma : (5 \times A)$, lines are the 5 high level component, the columns are the A substances

$\Sigma_j \gamma_{ij} = 1$, for i each of the high level components i = 1,...,5 and j = 1,..., A ; with A the number total of substances

Linear regression: % of the mass of each substance inside each high level component

```python
def substance_mass_matrices(dfs_clean):
    results = {}

    for socket in ['LGA', 'BGA']:
        socket_df = pd.concat([df[df['Socket'] == socket] for df in dfs_clean], ignore_index=True)
        if socket_df.empty:
            continue

        high_level_materials = socket_df['High Level Material'].unique()
        substances = socket_df['Substances'].unique()

        percentage_lists = {material: {sub: [] for sub in substances} for material in high_level_materials}

        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    percentage_lists[material][sub].append(percentage)

        # Initialize matrices
        percentage_matrix = pd.DataFrame(index=high_level_materials, columns=substances)
        std_dev_matrix = pd.DataFrame(index=high_level_materials, columns=substances)

        for material in high_level_materials:
            for sub in substances:
                percentages = percentage_lists[material][sub]
                if percentages:
                    percentage_matrix.loc[material, sub] = np.mean(percentages)
                    std_dev_matrix.loc[material, sub] = np.std(percentages)
                else:
                    percentage_matrix.loc[material, sub] = 0
                    std_dev_matrix.loc[material, sub] = 0

        results[socket] = {
            'Percentage': percentage_matrix.fillna(0),
            'Std Dev': std_dev_matrix.fillna(0)
        }

    return results
```

## **Calculation of the model**

Inputs of the model (parameters) : 
- total CPU surface [mm $^2$] = $s$, (1 $\times$ 1)
- total CPU mass [mg] = $w$, (1 $\times$ 1)

Other inputs of the model and intermediate results :
- high level components mass [mg] = $ \vec c $, $c_i$ with i = 1, 2, ... 5 ; (1 $\times$ 5). Calculated using the total mass $w$ using the function $g$

Output of the model :
- substances mass [mg] = $\vec m  = \vec c \times \vec \gamma $ ; are the masses of each of the susbtances calculated using the function $h$

 (1 $\times$ 5) $\times$ (5 $\times$ A) = (1 $\times$ A) with A the number total of substances

*Formated to be use in the Resilio API with the corresponding Ecoinvent elementary flow*

### 1. Function g

Input :
- total CPU mass [mg] $w$ : (1 $\times$ 1)
- percentages of the total mass of the CPU for each high level component [%], $\vec \beta$ : (1 $\times$ 5),

Output :
- mass of each of the high level components [mg], $\vec c$ : (1 $\times$ 5)

$g(w, \vec \beta) = \vec c = \vec \beta \times \vec w  <=> c_i = \beta_i \times w_i $  with i = 1, 2, 3, 4, 5

```python
def g(w, beta):
    """Calculate the mass of each high-level component."""
    c = np.zeros(5)
    for i in range(0, 4):
        c[i] = beta[i] * w
    return c
```

### 2. Function h

Input :
- mass of each of the high level components [mg], $\vec c$ : (1 $\times$ 5)
- percentage of mass of each substance in each high level component [%], $\vec \gamma$ : (5 $\times$ A), A is the number of substances

Output :
- mass of each of the substances [mg], $\vec m$ : (1 $\times$ A), A is the number of substances

$h(\vec c, \vec \gamma) = \vec m = \vec c \times \vec \gamma$

```python
def h(c_vec, gamma_vec):
    """Calculate the mass of each of the substances."""
    return c_vec @ gamma_vec
```

### Modelisation

```python
def get_beta(dfs_clean):
    """Get beta percentages from high_level_mass_matrix function."""
    beta_results = high_level_mass_matrix(dfs_clean)
    beta = {}
    for socket in beta_results:
        # Extract the percentage values directly from the DataFrame
        beta[socket] = beta_results[socket]['Percentage'].values.flatten()
    return beta


def get_gamma(dfs_clean):
    """Get gamma percentages from substance_mass_matrices function."""
    gamma_results = substance_mass_matrices(dfs_clean)
    gamma = {}
    for socket in gamma_results:
        gamma[socket] = {
            'matrix': gamma_results[socket]['Percentage'].to_numpy(),
            'substances': gamma_results[socket]['Percentage'].columns.tolist()
        }
    return gamma
```

```python
def model(total_cpu_mass, dfs_clean=None, beta=None, gamma=None, socket_type='LGA'):
    """
    Calculate the mass of each substance using the model, grouped by high-level material.

    Parameters:
    - total_cpu_mass: Total CPU mass in mg (scalar).
    - dfs_clean: List of DataFrames containing the data (optional if beta and gamma are provided).
    - beta: Precomputed beta (optional).
    - gamma: Precomputed gamma (DataFrame or dict, optional).
    - socket_type: Socket type (default is 'LGA').

    Returns:
    - DataFrame with columns: 'High Level', 'Substance', 'Mass [mg]'
    """
    if beta is None or gamma is None:
        if dfs_clean is None:
            raise ValueError("Either dfs_clean or both beta and gamma must be provided.")

        if beta is None:
            beta = get_beta(dfs_clean)[socket_type]

        if gamma is None:
            gamma_data = get_gamma(dfs_clean)[socket_type]
            gamma_matrix = gamma_data['matrix']
            substances = gamma_data['substances']
    else:
        if isinstance(gamma, dict):
            if 'matrix' in gamma:
                gamma_matrix = gamma['matrix']
                substances = gamma['substances']
            else:
                gamma_values = list(gamma.values())
                gamma_matrix = np.array(gamma_values).reshape(1, -1)
                substances = list(gamma.keys())
        else:  # Assume DataFrame
            gamma_matrix = gamma.values
            substances = gamma.columns

    # Normalize gamma_matrix to ensure each column sums to 1
    gamma_sum = gamma_matrix.sum(axis=0, keepdims=True)
    gamma_sum[gamma_sum == 0] = 1  # Avoid division by zero
    gamma_matrix = gamma_matrix / gamma_sum

    # Debugging prints
    print("Beta:", beta)
    print("Gamma Matrix:", gamma_matrix)
    print("Substances:", substances)

    # Calculate c_vec (mass of each high-level component)
    c_vec = g(total_cpu_mass, beta)
    print("c_vec:", c_vec)

    # Calculate substance masses per high-level component
    substance_masses_per_high_level = np.outer(c_vec, gamma_matrix)
    print("Substance Masses per High-Level:", substance_masses_per_high_level)

    # Flatten and create result DataFrame
    rows = []
    high_level_materials = ['HighLevel1', 'HighLevel2']  # Placeholder for high-level materials

    for i, material in enumerate(high_level_materials[:len(c_vec)]):
        for j, sub in enumerate(substances):
            mass = substance_masses_per_high_level[i, j]
            rows.append({
                'High Level': material,
                'Substance': sub,
                'Mass [mg]': mass
            })

    result_df = pd.DataFrame(rows)
    return result_df
```

# **Error propagation in the model**

### **1. Model definition**

Steps of the model :

- $\vec c = g(w, \vec\beta) = \vec\beta \times w → c_i = \beta_i \times w$

- $m = h(\vec c, \gamma) = \vec c \times \gamma → m_j = \Sigma_i (c_i \times\gamma_{ij}) = \Sigma_i (\beta_i \times w \times \gamma_{ij})$

Definition of the mass of a substance j : $m_j = w \times \Sigma_i (\vec\beta_i \times γ_{ij})$

Incertainties :

- $\sigma_w$ (uncertainty of total mass w) ?
- $\sigma_{\beta_i}$ (uncertainty of each $\beta_i$)
- $\sigma_{\gamma_{ij}}$ (uncertainty of each $\gamma_{ij}$)

Assumptions : Independent variables (but $\gamma$ and $\beta$ are correlated since they both have been estimated from the same dataset)

### **2. Error propagation**

The general formula for variance propagation for a function $y = f(x_1, x_2, ...)$ is : 

$\sigma_y^2 = \Sigma_i (\frac{∂ f}{∂ x_i})^2 \times \sigma_{x_i}^2$

In our case : $m_j = f(\vec \beta, w, \gamma) = w \times \Sigma_i (\beta_i \times \gamma_{ij})$ thus,

$\sigma_{m_j}^2 = (\frac{∂m_j}{∂w})² σ_w² + Σᵢ (\frac{∂m_j}{∂βᵢ})² σ_{βᵢ}² + Σᵢ (\frac{∂m_j}{∂γᵢⱼ})² σ_{γᵢⱼ}² $

With :

- $∂m_j/∂w = Σᵢ (βᵢ × γᵢⱼ)$
- $∂m_j/∂βᵢ = w × γᵢⱼ$
- $∂m_j/∂γᵢⱼ = w × βᵢ$


**Law of variance propagation : $K_{yy} = F · K_{xx} · F^T$**

With:

- $K_{yy}$ = covariance matrix of outputs
- F = Jacobian matrix of partial derivatives
- $K_{xx}$ = covariance matrix of inputs



```python
def propagate_error(model_output_df, total_cpu_mass, sigma_w, beta_df, gamma_df, sigma_gamma_df, cov_beta_gamma=None):
    """
    Propagates uncertainty through the model:
        m_j = w * sum_i (beta_i * gamma_ij)

    Explicitly constructs the Jacobian matrix F and the covariance matrix K_xx,
    accounting for potential correlations between beta and gamma.

    Parameters:
    - model_output_df: DataFrame from model() with columns ['High Level', 'Substance', 'Mass [mg]']
    - total_cpu_mass: scalar, total CPU mass in mg
    - sigma_w: scalar, uncertainty of total_cpu_mass
    - beta_df: DataFrame with 'Percentage' and 'Std Dev' for high-level components
    - gamma_df: DataFrame with substance fractions (n_beta x A)
    - sigma_gamma_df: DataFrame with std dev of gamma (n_beta x A)
    - cov_beta_gamma: Covariance matrix between beta and gamma (optional, default is None)

    Returns:
    - model_output_df with added columns:
        'Uncertainty [mg]', 'Rel Uncertainty [%]'
    """

    # Extract beta and sigma_beta
    beta = beta_df['Percentage'].values  # shape (n_beta,)
    sigma_beta = beta_df['Std Dev'].values  # shape (n_beta,)

    # Extract gamma and sigma_gamma
    gamma = gamma_df.values  # shape (n_beta, A)
    sigma_gamma = sigma_gamma_df.values  # shape (n_beta, A)

    # Number of substances and high-level components
    A = gamma.shape[1]
    n_beta = beta.shape[0]

    # Get unique high-level materials (must match beta order)
    high_level_materials = beta_df.index.tolist() if hasattr(beta_df, 'index') else [f"HL{i}" for i in range(n_beta)]

    # Initialize uncertainty array — now we need one per (high-level, substance) pair
    # We'll compute it per high-level material and substance
    sigma_m = np.zeros(len(model_output_df))  # shape (500,)

    # For each row in model_output_df (each (high-level, substance) pair)
    for idx, row in model_output_df.iterrows():
        high_level = row['High Level']
        substance = row['Substance']

        # Get index of high-level material
        try:
            i = high_level_materials.index(high_level)
        except ValueError:
            # If high-level not in beta, skip or set 0
            sigma_m[idx] = 0.0
            continue

        # Get index of substance
        try:
            j = gamma_df.columns.get_loc(substance)
        except KeyError:
            sigma_m[idx] = 0.0
            continue

        # Construct Jacobian matrix F for this (i, j) pair
        dmdw = beta[i] * gamma[i, j]  # ∂m_ij/∂w
        dmdbeta = np.zeros(n_beta)
        dmdbeta[i] = total_cpu_mass * gamma[i, j]  # ∂m_ij/∂β_i
        dmdgamma = np.zeros(n_beta)
        dmdgamma[i] = total_cpu_mass * beta[i]  # ∂m_ij/∂γ_ij

        # Combine into F (1 x (1 + n_beta + n_beta) row vector)
        F = np.hstack([[dmdw], dmdbeta, dmdgamma])

        # Construct covariance matrix K_xx
        # Diagonal elements for w, beta, and gamma
        diag_elements = np.hstack([[sigma_w**2], sigma_beta**2, sigma_gamma[:, j]**2])

        # Initialize K_xx as a diagonal matrix
        K_xx = np.diag(diag_elements)

        # If cov_beta_gamma is provided, update the off-diagonal elements
        if cov_beta_gamma is not None:
            # Assume cov_beta_gamma is shape (n_beta, n_beta)
            for k in range(n_beta):
                for l in range(n_beta):
                    # Update covariance between beta_k and gamma_lj
                    K_xx[1 + k, 1 + n_beta + l] = cov_beta_gamma[k, l]
                    K_xx[1 + n_beta + l, 1 + k] = cov_beta_gamma[k, l]

        # Compute variance using F * K_xx * F^T
        sigma_m_sq = F @ K_xx @ F.T
        sigma_m[idx] = np.sqrt(sigma_m_sq)

    # Add to DataFrame
    result_df = model_output_df.copy()
    result_df['Uncertainty [mg]'] = sigma_m
    result_df['Rel Uncertainty [%]'] = (sigma_m / result_df['Mass [mg]']) * 100

    return result_df
```

```python
def propagate_error_cov(model_output_df, total_cpu_mass, sigma_w, beta_df, gamma_df, sigma_gamma_df, cov_beta=None, cov_gamma=None, cov_beta_gamma=None):
    """
    Propagates uncertainty through the model using covariance matrices.

    Computes two bounds:
    - Underestimate: assumes no correlations (diagonal covariances)
    - Overestimate: uses provided covariances (if any)

    Parameters:
    - model_output_df: DataFrame with columns ['High Level', 'Substance', 'Mass [mg]']
    - total_cpu_mass: scalar, total CPU mass in mg
    - sigma_w: scalar, uncertainty of total_cpu_mass
    - beta_df: DataFrame with 'Percentage' and 'Std Dev' for high-level components
    - gamma_df: DataFrame with substance fractions (n_beta x A)
    - sigma_gamma_df: DataFrame with std dev of gamma (n_beta x A)
    - cov_beta: Covariance matrix for beta (n_beta x n_beta) (optional)
    - cov_gamma: Covariance matrix for gamma (n_beta x n_beta) (optional)
    - cov_beta_gamma: Covariance matrix between beta and gamma (optional, n_beta x n_beta)

    Returns:
    - model_output_df with added columns:
        'Sigma [mg] (Underestimate)', 'Sigma [%] (Underestimate)',
        'Sigma [mg] (Overestimate)', 'Sigma [%] (Overestimate)'
    """

    # Extract beta and sigma_beta
    beta = beta_df['Percentage'].values  # shape (n_beta,)
    sigma_beta = beta_df['Std Dev'].values  # shape (n_beta,)

    # Extract gamma and sigma_gamma
    gamma = gamma_df.values  # shape (n_beta, A)
    sigma_gamma = sigma_gamma_df.values  # shape (n_beta, A)

    # Number of substances and high-level components
    A = gamma.shape[1]
    n_beta = beta.shape[0]

    # Get high-level material names (must match beta order)
    high_level_materials = beta_df.index.tolist() if hasattr(beta_df, 'index') else [f"HL{i}" for i in range(n_beta)]

    # Initialize uncertainty arrays
    sigma_m_under = np.zeros(len(model_output_df))
    sigma_m_over = np.zeros(len(model_output_df))

    # For each row in model_output_df
    for idx, row in model_output_df.iterrows():
        high_level = row['High Level']
        substance = row['Substance']

        # Get index of high-level material
        try:
            i = high_level_materials.index(high_level)
        except ValueError:
            sigma_m_under[idx] = 0.0
            sigma_m_over[idx] = 0.0
            continue

        # Get index of substance
        try:
            j = gamma_df.columns.get_loc(substance)
        except KeyError:
            sigma_m_under[idx] = 0.0
            sigma_m_over[idx] = 0.0
            continue

        # Construct Jacobian matrix F for this (i, j) pair
        dmdw = beta[i] * gamma[i, j]  # ∂m_ij/∂w
        dmdbeta = np.zeros(n_beta)
        dmdbeta[i] = total_cpu_mass * gamma[i, j]  # ∂m_ij/∂β_i
        dmdgamma = np.zeros(n_beta)
        dmdgamma[i] = total_cpu_mass * beta[i]  # ∂m_ij/∂γ_ij

        # Combine into F (1 x (1 + n_beta + n_beta) row vector)
        F = np.hstack([[dmdw], dmdbeta, dmdgamma])

        # ————————————————————————————————————————————————————————————————
        # Underestimate: Assume no correlations (diagonal)
        # ————————————————————————————————————————————————————————————————

        K_w = np.array([[sigma_w**2]])
        K_beta_under = np.diag(sigma_beta**2)
        K_gamma_under = np.diag(sigma_gamma[:, j]**2)
        K_beta_gamma_under = np.zeros((n_beta, n_beta))

        K_xx_under = np.block([
            [K_w, np.zeros((1, n_beta)), np.zeros((1, n_beta))],
            [np.zeros((n_beta, 1)), K_beta_under, K_beta_gamma_under],
            [np.zeros((n_beta, 1)), K_beta_gamma_under.T, K_gamma_under]
        ])

        sigma_m_sq_under = F @ K_xx_under @ F.T
        sigma_m_under[idx] = np.sqrt(sigma_m_sq_under)

        # ————————————————————————————————————————————————————————————————
        # Overestimate: Use provided covariances (if any)
        # ————————————————————————————————————————————————————————————————

        # Use provided covariances if available, else fall back to diagonal
        K_beta_over = cov_beta if cov_beta is not None else np.diag(sigma_beta**2)
        K_gamma_over = cov_gamma if cov_gamma is not None else np.diag(sigma_gamma[:, j]**2)
        K_beta_gamma_over = cov_beta_gamma if cov_beta_gamma is not None else np.zeros((n_beta, n_beta))

        K_xx_over = np.block([
            [K_w, np.zeros((1, n_beta)), np.zeros((1, n_beta))],
            [np.zeros((n_beta, 1)), K_beta_over, K_beta_gamma_over],
            [np.zeros((n_beta, 1)), K_beta_gamma_over.T, K_gamma_over]
        ])

        sigma_m_sq_over = F @ K_xx_over @ F.T
        sigma_m_over[idx] = np.sqrt(sigma_m_sq_over)

    # Add to DataFrame
    result_df = model_output_df.copy()
    result_df['Sigma [mg] (Underestimate)'] = sigma_m_under
    result_df['Sigma [%] (Underestimate)'] = (sigma_m_under / result_df['Mass [mg]']) * 100
    result_df['Sigma [mg] (Overestimate)'] = sigma_m_over
    result_df['Sigma [%] (Overestimate)'] = (sigma_m_over / result_df['Mass [mg]']) * 100

    return result_df
```

## **Workflow**

For a set of observations $l$ and a model that predicts values $f(x)$, the residuals $\hat v$ are defined as : 

$\hat v = l - f(\hat x)$, with $f(\hat x)$ the vector of predicted values based on the estimated parameters $\hat x$

### **1. Data preparation**

Inputs : 10 substances, 7 samples, the percentage and the error for each of the substances in each of the samples

- Observation vector $\vec l : (7 \times 10)$ contains the measurement for each of the 10 substances for each of the 7 samples
- Covariance matrix $K_{ll} : (70 \times 70)$ diagonal matrix with variances of observation $(7 \times 7)$ per substance
- Weight matrix $P : (70 \times 70)$ inverse of the covariance matrix

 $P=K_{ll}^{-1}$

```python
def prepare_data(cpudata_substances):
    """
    Prepare data for calibration.

    Args:
        cpudata_substances: DataFrame containing measurements and uncertainties for each substance.

    Returns:
        l: Observation vector (7x10)
        K_ll: Covariance matrix (70x70)
        P: Weight matrix (70x70)
    """
    l = cpudata_substances.iloc[:, 2:12].values  # Observation vector (7x10)
    sigma_l = cpudata_substances.iloc[:, 12:22].values  # Standard deviations (7x10)

    # Construct covariance matrix K_ll (70x70)
    K_ll_diag = sigma_l.flatten() ** 2
    K_ll_diag[K_ll_diag == 0] = 1e-10  # Avoid division by zero
    K_ll = np.diag(K_ll_diag)

    # Construct weight matrix P (70x70)
    P = np.diag(1 / K_ll_diag)

    return l, K_ll, P
```

### **2. Calculate $\beta$ parameters**

Calculation of $\beta$ using the least squares method:

- Design matrix $A : (7 \times 10)$ contains the measurements for each substance across the 7 samples
- Observation vector $\vec y : (7 \times 1)$ contains the depend variable, weight $w$

$\beta = (A^T P A)^{-1}A^T P y$ : $(10 \times 1)$

In least squares estimation, the goal is to minimize the sum of the squared residuals, which helps in estimating the model parameters.

```python
def calculate_beta(dfs_clean):
    """
    Calculate beta parameters using the least squares method.

    Args:
        dfs_clean: DataFrame containing measurements and uncertainties.

    Returns:
        beta_dict: Dictionary of beta parameters for each CPU.
    """
    dfs = dfs_clean.copy()
    dfs['w'] = 100 / dfs['w']

    substances = ['Al', 'Co', 'Cr', 'Cu', 'Fe', 'Mg', 'Mn', 'Ni', 'Zn', 'Sn']
    beta_dict = {}

    for cpu in dfs['CPU'].unique():
        cpu_df = dfs[dfs['CPU'] == cpu]
        A = cpu_df[substances].values  # Design matrix (7x10)
        y = cpu_df['w'].values  # Observation vector (7x1)

        # Weight matrix P (7x7)
        P = np.diag(1 / (cpu_df['sigma_w'] ** 2))

        # Calculate beta with regularization to handle potential singularity
        A_T_P_A = A.T @ P @ A
        regularization = 1e-6 * np.eye(A_T_P_A.shape[0])
        A_T_P_A_reg = A_T_P_A + regularization
        A_T_P_y = A.T @ P @ y
        beta = np.linalg.solve(A_T_P_A_reg, A_T_P_y)

        beta_dict[cpu] = beta

    return beta_dict

```


### **3. Calculate $\gamma$ parameters**

Calibrate the $\gamma$ parameters for each substance using the parametric compensation model:

$\gamma_j = (A^T P A)^{-1}A^T P\space l$ : scalar $(1\times1)$ 

```python
def calibrate_gamma(l, P, beta, substance, m_j, sigma_mj, substances):
    """
    Calibrate gamma parameters for a specific substance.

    Args:
        l: Observation vector (7x10)
        P: Weight matrix (70x70)
        beta: Beta parameters (10x1)
        substance: Name of the substance
        m_j: Measurements for the substance (7x1)
        sigma_mj: Standard deviations for the substance (7x1)
        substances: List of substances

    Returns:
        gamma_j: Gamma parameter for the substance (scalar)
        residuals: Residuals for the substance (7x1)
        Q_gamma_j_gamma_j: Cofactor matrix for gamma (scalar)
        z_i: Redundancy numbers (7x1)
    """
    if all(m_j == 0):
        return 0, np.zeros_like(m_j), 0, np.zeros_like(m_j)

    # Design matrix A (7x1)
    w = l[:, 0]  # Assuming 'w' is the first column in l
    beta_idx = substances.index(substance)
    A = np.outer(w, beta[beta_idx])

    # Weight matrix P for the substance (7x7)
    P_j = np.diag(1 / sigma_mj ** 2)

    # Calculate gamma_j with regularization to handle potential singularity
    A_T_P_A = A.T @ P_j @ A
    regularization = 1e-6 * np.eye(A_T_P_A.shape[0])
    A_T_P_A_reg = A_T_P_A + regularization
    A_T_P_l = A.T @ P_j @ m_j

    gamma_j = np.linalg.solve(A_T_P_A_reg, A_T_P_l)

    # Calculate residuals
    residuals = m_j - A @ gamma_j

    # Cofactor matrix Q_gamma_j_gamma_j (scalar)
    Q_gamma_j_gamma_j = np.linalg.inv(A_T_P_A_reg)

    # Cofactor matrix Q_ll (7x7)
    Q_ll = np.diag(1 / sigma_mj ** 2)

    # Cofactor matrix of residuals Q_vv (7x7)
    Q_vv = Q_ll - A @ Q_gamma_j_gamma_j @ A.T

    # Redundancy numbers z_i (7x1)
    z_i = np.diag(Q_vv @ P_j)

    return gamma_j, residuals, Q_gamma_j_gamma_j, z_i
```



### **4. Analyze results**

Analyze the results of the calibration, including the calculation of standardized residuals and outlier detection:

- Residuals vector $\hat v : (7\times 1)$ contaisn the residuals for a specific substances for each sample
- Cofactor matrix $Q_{\gamma_j \gamma_j} = (A^T P A)^{-1} $ : scalar $(1\times 1)$ inverse of the normal matrix $(A^T P A)$
- Cofactor matrix $Q_{l l} = \frac{K_{ll}}{\sigma^2_0} $ : $(7\times 7)$ derived from the covariance matrix by factoring out the variance of unit weight
- Cofactor matrix of residuals $Q_{\hat v \hat v} $ : $(7\times 7)$ 

$Q_{\hat v \hat v} = Q_{l l} - A Q_{\gamma_j \gamma_j} A^T$

- Redundency numbers $z_i$ : $(7 \times 1)$, diagonal elements of $Q_{\hat v \hat v}P$

- Standardized residuals $w_i$ : $(7 \times 1)$, contains the standardized residuals for a specific substance

$w_i = \frac {\hat v_i}{\sqrt z_i \times \hat\sigma _0}$

```python
def analyze_results(residuals, z_i, sigma_0_hat):
    """
    Analyze results of the calibration.

    Args:
        residuals: Residuals for the substance (7x1)
        z_i: Redundancy numbers (7x1)
        sigma_0_hat: A posteriori standard deviation (scalar)

    Returns:
        w_i: Standardized residuals (7x1)
        outlier: Boolean indicating if an outlier is detected
    """
    # Standardized residuals w_i (7x1)
    w_i = residuals / (np.sqrt(z_i) * sigma_0_hat)

    # Detect outliers
    outlier = np.any(np.abs(w_i) > 2.5)

    return w_i, outlier
```

### **5. Correction of the model**

Correct and update the model parameters based on the residuals:

$\beta _{updated} = \beta - learning_rate \times \delta\beta _i$

- $\beta$ residuals $\delta\beta_i : (10 \times 1)$ contains the mean residuals for each $\beta$ parameter
- Updated $\beta_{updated} : (10\times 1)$ contains the updated $\beta$ parameters
- Updated $\gamma_{updated}$ : scalar for each substance $(1\times 1)$, contaisn the updated $\gamma$ parameters

```python
def correct_model(beta, gamma_dict, residuals_dict, learning_rate=0.1):
    """
    Correct and update the model parameters based on the residuals.

    Args:
        beta: Beta parameters (10x1)
        gamma_dict: Dictionary of gamma parameters for each substance
        residuals_dict: Dictionary of residuals for each substance
        learning_rate: Learning rate for updating parameters

    Returns:
        beta_updated: Updated beta parameters (10x1)
        gamma_updated_dict: Updated gamma parameters for each substance
    """
    # Calculate beta residuals (10x1)
    delta_beta = np.zeros_like(beta)
    for i in range(len(beta)):
        for substance, residuals in residuals_dict.items():
            delta_beta[i] += np.mean(residuals)

    # Update beta
    beta_updated = beta - learning_rate * delta_beta / len(residuals_dict)

    # Normalize beta_updated to ensure their sum is 1
    beta_updated = beta_updated / np.sum(beta_updated)

    # Update gamma
    gamma_updated_dict = {}
    for substance, gamma in gamma_dict.items():
        residuals = residuals_dict[substance]
        gamma_updated = gamma - learning_rate * np.mean(residuals)
        gamma_updated_dict[substance] = gamma_updated

    return beta_updated, gamma_updated_dict


```

```python
def main_workflow(cpudata_substances, dfs_clean):
    """
    Main workflow for calibration and analysis.

    Args:
        cpudata_substances: DataFrame containing measurements and uncertainties for each substance.
        dfs_clean: DataFrame containing measurements and uncertainties for beta calculation.

    Returns:
        beta: Beta parameters (10x1)
        gamma_dict: Dictionary of gamma parameters for each substance
        analysis_results: Dictionary of analysis results for each substance
        beta_updated: Updated beta parameters (10x1)
        gamma_updated_dict: Updated gamma parameters for each substance
    """
    # Step 1: Data Preparation
    l = cpudata_substances.iloc[:, 2:12].values  # Observation vector (7x10)
    sigma_l = cpudata_substances.iloc[:, 12:22].values  # Standard deviations (7x10)

    # Construct covariance matrix K_ll (70x70)
    K_ll = np.diag(sigma_l.flatten() ** 2)

    # Construct weight matrix P (70x70)
    P = np.diag(1 / np.diag(K_ll))

    # Step 2: Calculate Beta Parameters
    beta_dict = calculate_beta(dfs_clean)
    beta = beta_dict['LGA']  # Assuming 'LGA' is the CPU of interest

    # Step 3: Calculate Gamma Parameters
    gamma_dict = {}
    residuals_dict = {}
    Q_gamma_dict = {}
    z_dict = {}

    substances = ['Al', 'Co', 'Cr', 'Cu', 'Fe', 'Mg', 'Mn', 'Ni', 'Zn', 'Sn']

    for substance in substances:
        m_j = cpudata_substances[substance].values
        sigma_mj = cpudata_substances[f'sigma_{substance}'].values

        gamma_j, residuals, Q_gamma_j_gamma_j, z_i = calibrate_gamma(l, P, beta, substance, m_j, sigma_mj, substances)

        gamma_dict[substance] = gamma_j
        residuals_dict[substance] = residuals
        Q_gamma_dict[substance] = Q_gamma_j_gamma_j
        z_dict[substance] = z_i

    # Step 4: Analyze Results
    analysis_results = {}
    for substance in substances:
        residuals = residuals_dict[substance]
        z_i = z_dict[substance]
        sigma_0_hat = np.sqrt(np.sum(residuals ** 2) / (len(residuals) - 1)) if len(residuals) > 1 else 0  # A posteriori standard deviation

        if sigma_0_hat > 0:
            w_i, outlier = analyze_results(residuals, z_i, sigma_0_hat)
        else:
            w_i, outlier = np.zeros_like(residuals), False

        analysis_results[substance] = {'w_i': w_i, 'outlier': outlier}

    # Step 5: Correction of the Model
    beta_updated, gamma_updated_dict = correct_model(beta, gamma_dict, residuals_dict)

    return beta, gamma_dict, analysis_results, beta_updated, gamma_updated_dict
```

```python
def compare_substance_quantities(cpudata_substances, dfs_clean, total_cpu_mass=2800):
    """
    Compare substance quantities before and after model correction.

    Args:
        cpudata_substances: DataFrame containing measurements and uncertainties for each substance.
        dfs_clean: DataFrame containing measurements and uncertainties for beta calculation.
        total_cpu_mass: Total CPU mass in mg.

    Returns:
        before_correction: DataFrame with substance quantities before correction.
        after_correction: DataFrame with substance quantities after correction.
    """
    # Step 1: Run the model initially
    beta, gamma_dict, analysis_results, beta_updated, gamma_updated_dict = main_workflow(cpudata_substances, dfs_clean)

    # Convert gamma_dict to a matrix format
    gamma_values = list(gamma_dict.values())
    gamma_before_matrix = np.array(gamma_values).reshape(1, -1)
    gamma_before_matrix = gamma_before_matrix / gamma_before_matrix.sum(axis=1, keepdims=True)

    # Create gamma_before dictionary in the expected format
    gamma_before = {'matrix': gamma_before_matrix, 'substances': list(gamma_dict.keys())}

    # Calculate initial substance quantities
    before_correction = model(total_cpu_mass, dfs_clean, beta=beta, gamma=gamma_before, socket_type='LGA')

    # Step 2: Update the model
    # Update the data with the new gamma parameters
    for substance in gamma_updated_dict:
        if substance in cpudata_substances.columns:
            cpudata_substances[substance] = gamma_updated_dict[substance] * cpudata_substances['w'] / 100

    # Convert gamma_updated_dict to a matrix format
    gamma_updated_values = list(gamma_updated_dict.values())
    gamma_after_matrix = np.array(gamma_updated_values).reshape(1, -1)
    gamma_after_matrix = gamma_after_matrix / gamma_after_matrix.sum(axis=1, keepdims=True)

    # Create gamma_after dictionary in the expected format
    gamma_after = {'matrix': gamma_after_matrix, 'substances': list(gamma_updated_dict.keys())}

    # Rerun the model with updated data
    beta, gamma_dict, analysis_results, beta_updated, gamma_updated_dict = main_workflow(cpudata_substances, dfs_clean)

    # Calculate updated substance quantities
    after_correction = model(total_cpu_mass, dfs_clean, beta=beta_updated, gamma=gamma_after, socket_type='LGA')

    return before_correction, after_correction
```

```python
def print_comparison(before_correction, after_correction):
    """
    Print the comparison of substance quantities before and after correction.

    Args:
        before_correction: DataFrame with substance quantities before correction.
        after_correction: DataFrame with substance quantities after correction.
    """
    print("Substance Quantities Before Correction (mg):")
    print(before_correction)

    print("\nSubstance Quantities After Correction (mg):")
    print(after_correction)

    # Calculate and print the change in substance quantities
    merged_df = pd.merge(before_correction, after_correction, on=['High Level', 'Substance'], suffixes=('_Before', '_After'))
    merged_df['Change [mg]'] = merged_df['Mass [mg]_After'] - merged_df['Mass [mg]_Before']

    print("\nChange in Substance Quantities (mg):")
    print(merged_df[['High Level', 'Substance', 'Change [mg]']])
```

```python
# Run and compare the substance quantities
before_correction, after_correction = compare_substance_quantities(cpudata_substances, dfs_clean, total_cpu_mass=2800)

# Print the comparison
print_comparison(before_correction, after_correction)
```
