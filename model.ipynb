```python
# Folder containing Excel files
folder_path = r"C:\Users\Arthur Resilio\Desktop\Excel_cleaned"

# List all Excel files
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".xlsx")]

dfs = []

for file in all_files:
    try:
        df = pd.read_excel(file)

        # --- 1. Clean column names ---
        df.columns = (
            df.columns.astype(str)
            .str.strip()
            .str.replace('\n', ' ', regex=False)
            .str.replace('\r', '', regex=False)
            .str.replace('\xa0', ' ', regex=False)
        )

        # --- 2. Identify date columns (keep as datetime) ---
        date_columns = [col for col in df.columns if 'date' in col.lower()]
        for col in date_columns:
            try:
                df[col] = pd.to_datetime(df[col], errors='ignore')
            except:
                pass

        # --- 3. Convert mass/weight columns to numeric ---
        mass_columns = ['Mass [mg]', 'Weight [mg]', 'Mass\n[mg]', 'Weight\n[mg]']
        for col in mass_columns:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

        dfs.append(df)

    except Exception as e:
        print(f"Error loading {file}: {e}")

print(f"✅ Total datasets successfully loaded: {len(dfs)}")

# Optionally save dfs to a pickle for use in the next script
pd.to_pickle(dfs, r"C:\Users\Arthur Resilio\Desktop\dfs_loaded.pkl")

dfs = pd.read_pickle(r"C:\Users\Arthur Resilio\Desktop\dfs_loaded.pkl")

def clean_name(name):
    """Standardize text: lowercase, strip spaces, remove parentheses contents."""
    if not isinstance(name, str):
        return name
    name = name.strip()
    name = re.sub(r'\s+', ' ', name)  # collapse multiple spaces
    name = name.lower()
    name = re.sub(r'\s*\(.*?\)\s*', '', name)  # remove parentheses and contents
    return name

def compute_surface_from_first_two(size_series):
    """
    Compute surface using the first two valid numeric Size [mm] values for a given CPU.
    If only one value exists, assume square.
    """
    valid_sizes = pd.to_numeric(size_series, errors="coerce").dropna().values
    if len(valid_sizes) >= 2:
        return valid_sizes[0] * valid_sizes[1]
    elif len(valid_sizes) == 1:
        return valid_sizes[0] ** 2
    else:
        return None

def clean_bom_df(df):
    """
    Cleans a single CPU BOM DataFrame:
    - Forward fill CPU metadata (Name, Socket, Size, Date)
    - Standardize names and substances
    - Compute total mass and surface area per CPU
    """
    df = df.copy()

    # Forward fill CPU metadata
    for col in ['Name', 'Socket', 'Size [mm]', 'Date']:
        if col in df.columns:
            df[col] = df[col].ffill()

    # Standardize text columns
    for col in ['Homogeneous Material', 'Substances']:
        if col in df.columns:
            df[col] = df[col].apply(clean_name)

    # Convert mass column to numeric
    if 'Mass [mg]' in df.columns:
        df['Mass [mg]'] = pd.to_numeric(df['Mass [mg]'], errors='coerce').fillna(0)
    else:
        df['Mass [mg]'] = 0

    # ✅ Compute total CPU mass per Name
    df['CPU Total Mass [mg]'] = df.groupby('Name')['Mass [mg]'].transform('sum')

    # ✅ Compute surface per CPU from the first two valid size values
    df['Surface [mm²]'] = df.groupby('Name')['Size [mm]'].transform(compute_surface_from_first_two)

    return df

# Clean all DataFrames
dfs_high_level = [clean_bom_df(df) for df in dfs]

# Save cleaned DataFrames
output_path = r"C:\Users\Arthur Resilio\Desktop\dfs_cleaned.pkl"
pd.to_pickle(dfs_high_level, output_path)

```

## Helper function to calculate standard deviation from a list of percentages

```python
def calculate_stats(percentage_lists):
    """Calculate mean and standard deviation from a list of percentages."""
    stats = {}
    for key, percentages in percentage_lists.items():
        if percentages:
            stats[key] = {
                'Percentage': np.mean(percentages),
                'Std Dev': np.std(percentages)
            }
        else:
            stats[key] = {
                'Percentage': 0,
                'Std Dev': 0
            }
    return stats
```

## **Calculation of $\vec\beta$**

$\vec\beta : (5 \times 1)$

$\Sigma_i \beta_i = 1$, i = 1,2,3,4,5

Linear regression: % of the total mass of the CPU for each high level component


```python
def high_level_mass_matrix(dfs_clean):
    results = {}

    for socket in ['LGA', 'BGA']:
        percentage_lists = {material: [] for material in pd.concat(dfs_clean)['High Level Material'].unique()}

        for df in dfs_clean:
            socket_df = df[df['Socket'] == socket]
            if socket_df.empty:
                continue

            total_mass = socket_df['Mass [mg]'].sum()
            for material in percentage_lists.keys():
                material_mass = socket_df[socket_df['High Level Material'] == material]['Mass [mg]'].sum()
                percentage = material_mass / total_mass if total_mass else 0
                percentage_lists[material].append(percentage)

        stats = calculate_stats(percentage_lists)
        results[socket] = pd.DataFrame.from_dict(stats, orient='index')

    return results
```

## **Calculation of matrix **$\gamma$****

$\gamma : (5 \times A)$, lines are the 5 high level component, the columns are the A substances

$\Sigma_j \gamma_{ij} = 1$, for i each of the high level components i = 1,...,5 and j = 1,..., A ; with A the number total of substances

Linear regression: % of the mass of each substance inside each high level component

```python
def substance_mass_matrices(dfs_clean):
    results = {}

    for socket in ['LGA', 'BGA']:
        socket_df = pd.concat([df[df['Socket'] == socket] for df in dfs_clean], ignore_index=True)
        if socket_df.empty:
            continue

        high_level_materials = socket_df['High Level Material'].unique()
        substances = socket_df['Substances'].unique()

        percentage_lists = {material: {sub: [] for sub in substances} for material in high_level_materials}

        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    percentage_lists[material][sub].append(percentage)

        # Initialize matrices
        percentage_matrix = pd.DataFrame(index=high_level_materials, columns=substances)
        std_dev_matrix = pd.DataFrame(index=high_level_materials, columns=substances)

        for material in high_level_materials:
            for sub in substances:
                percentages = percentage_lists[material][sub]
                if percentages:
                    percentage_matrix.loc[material, sub] = np.mean(percentages)
                    std_dev_matrix.loc[material, sub] = np.std(percentages)
                else:
                    percentage_matrix.loc[material, sub] = 0
                    std_dev_matrix.loc[material, sub] = 0

        results[socket] = {
            'Percentage': percentage_matrix.fillna(0),
            'Std Dev': std_dev_matrix.fillna(0)
        }

    return results
```

## **Calculation of the model**

Inputs of the model (parameters) : 
- total CPU surface [mm $^2$] = $s$, (1 $\times$ 1)
- total CPU mass [mg] = $w$, (1 $\times$ 1)

Other inputs of the model and intermediate results :
- high level components mass [mg] = $ \vec c $, $c_i$ with i = 1, 2, ... 5 ; (1 $\times$ 5). Calculated using the total mass $w$ using the function $g$

Output of the model :
- substances mass [mg] = $\vec m  = \vec c \times \vec \gamma $ ; are the masses of each of the susbtances calculated using the function $h$

 (1 $\times$ 5) $\times$ (5 $\times$ A) = (1 $\times$ A) with A the number total of substances

*Formated to be use in the Resilio API with the corresponding Ecoinvent elementary flow*

### 1. Function g

Input :
- total CPU mass [mg] $w$ : (1 $\times$ 1)
- percentages of the total mass of the CPU for each high level component [%], $\vec \beta$ : (1 $\times$ 5),

Output :
- mass of each of the high level components [mg], $\vec c$ : (1 $\times$ 5)

$g(w, \vec \beta) = \vec c = \vec \beta \times \vec w  <=> c_i = \beta_i \times w_i $  with i = 1, 2, 3, 4, 5

```python
def g(w, beta):
    """Calculate the mass of each high-level component."""
    c = np.zeros(5)
    for i in range(0, 4):
        c[i] = beta[i] * w
    return c
```

### 2. Function h

Input :
- mass of each of the high level components [mg], $\vec c$ : (1 $\times$ 5)
- percentage of mass of each substance in each high level component [%], $\vec \gamma$ : (1 $\times$ A), A is the number of substances

Output :
- mass of each of the substances [mg], $\vec m$ : (1 $\times$ A), A is the number of substances

$h(\vec c, \vec \gamma) = \vec m = \vec c \times \vec \gamma$

```python
def h(c_vec, gamma_vec):
    """Calculate the mass of each of the substances."""
    return c_vec @ gamma_vec
```

### Modelisation

```python
def get_beta(dfs_clean):
    """Get beta percentages from high_level_mass_matrix function."""
    beta_results = high_level_mass_matrix(dfs_clean)
    beta = {}
    for socket in beta_results:
        # Extract the percentage values directly from the DataFrame
        beta[socket] = beta_results[socket]['Percentage'].values.flatten()
    return beta


def get_gamma(dfs_clean):
    """Get gamma percentages from substance_mass_matrices function."""
    gamma_results = substance_mass_matrices(dfs_clean)
    gamma = {}
    for socket in gamma_results:
        gamma[socket] = {
            'matrix': gamma_results[socket]['Percentage'].to_numpy(),
            'substances': gamma_results[socket]['Percentage'].columns.tolist()
        }
    return gamma
```

```python
def model(total_cpu_mass, dfs_clean, socket_type='LGA'):
    """
    Calculate the mass of each substance using the model.

    Parameters:
    - surface: Surface in mm² (scalar).
    - total_cpu_mass: Total CPU mass in mg (scalar).
    - dfs_clean: List of DataFrames containing the data.
    - material: Material to analyze (default is 'carrier').
    - socket_type: Socket type (default is 'LGA').

    Returns:
    - DataFrame with substance names and their masses.
    """
    # Get beta, and gamma
    beta = get_beta(dfs_clean)[socket_type]
    gamma_data = get_gamma(dfs_clean)[socket_type]
    substances = gamma_data['substances']
    gamma_vec = gamma_data['matrix']

    # Calculate c_vec using g
    c_vec = g(total_cpu_mass, beta)
    
    # Calculate substance masses using h
    substance_masses = h(c_vec, gamma_vec)

    # Create a DataFrame with substance names and their masses
    result_df = pd.DataFrame({'Substance': substances, 'Mass [mg]': substance_masses})
    return result_df

```

```python
def element_masses_to_api_text(substance_masses_df, DM=DIRECT_MAPPING):
    """
    Convert substance masses DataFrame to a formatted text message for API usage.
    Args:
        substance_masses_df: DataFrame with columns ['Substance', 'Mass [mg]'].
        DM: Dictionary for direct mapping of substances to terminal classes.
    Returns:
        Formatted string for API usage.
    """
    result = "SUBSTANCES: list[Exchange] = [\n"
    for _, row in substance_masses_df.iterrows():
        element = row['Substance']
        # Skip if element is not a string
        if not isinstance(element, str):
            continue
        mass = row['Mass [mg]']
        terminal_class = DM.get(element.lower())
        if terminal_class:
            mass_str = f"{mass:.15g}"
            result += f"    Exchange({mass_str} * 1e-6, {terminal_class}),\n"
    result += "]\n"
    return result

```
