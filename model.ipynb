```python
# Folder containing Excel files
folder_path = r"C:\Users\Arthur Resilio\Desktop\Excel_cleaned"

# List all Excel files
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".xlsx")]

dfs = []

for file in all_files:
    try:
        df = pd.read_excel(file)

        # --- 1. Clean column names ---
        df.columns = (
            df.columns.astype(str)
            .str.strip()
            .str.replace('\n', ' ', regex=False)
            .str.replace('\r', '', regex=False)
            .str.replace('\xa0', ' ', regex=False)
        )

        # --- 2. Identify date columns (keep as datetime) ---
        date_columns = [col for col in df.columns if 'date' in col.lower()]
        for col in date_columns:
            try:
                df[col] = pd.to_datetime(df[col], errors='ignore')
            except:
                pass

        # --- 3. Convert mass/weight columns to numeric ---
        mass_columns = ['Mass [mg]', 'Weight [mg]', 'Mass\n[mg]', 'Weight\n[mg]']
        for col in mass_columns:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

        dfs.append(df)

    except Exception as e:
        print(f"Error loading {file}: {e}")

print(f"Total datasets successfully loaded: {len(dfs)}")

# Optionally save dfs to a pickle for use in the next script
pd.to_pickle(dfs, r"C:\Users\Arthur Resilio\Desktop\dfs_loaded.pkl")

dfs = pd.read_pickle(r"C:\Users\Arthur Resilio\Desktop\dfs_loaded.pkl")

def clean_name(name):
    """Standardize text: lowercase, strip spaces, remove parentheses contents."""
    if not isinstance(name, str):
        return name
    name = name.strip()
    name = re.sub(r'\s+', ' ', name)  # collapse multiple spaces
    name = name.lower()
    name = re.sub(r'\s*\(.*?\)\s*', '', name)  # remove parentheses and contents
    return name

def compute_surface_from_first_two(size_series):
    """
    Compute surface using the first two valid numeric Size [mm] values for a given CPU.
    If only one value exists, assume square.
    """
    valid_sizes = pd.to_numeric(size_series, errors="coerce").dropna().values
    if len(valid_sizes) >= 2:
        return valid_sizes[0] * valid_sizes[1]
    elif len(valid_sizes) == 1:
        return valid_sizes[0] ** 2
    else:
        return None

def clean_bom_df(df):
    """
    Cleans a single CPU BOM DataFrame:
    - Forward fill CPU metadata (Name, Socket, Size, Date)
    - Standardize names and substances
    - Compute total mass and surface area per CPU
    """
    df = df.copy()

    # Forward fill CPU metadata
    for col in ['Name', 'Socket', 'Size [mm]', 'Date']:
        if col in df.columns:
            df[col] = df[col].ffill()

    # Standardize text columns
    for col in ['Homogeneous Material', 'Substances']:
        if col in df.columns:
            df[col] = df[col].apply(clean_name)

    # Convert mass column to numeric
    if 'Mass [mg]' in df.columns:
        df['Mass [mg]'] = pd.to_numeric(df['Mass [mg]'], errors='coerce').fillna(0)
    else:
        df['Mass [mg]'] = 0

    # Compute total CPU mass per Name
    df['CPU Total Mass [mg]'] = df.groupby('Name')['Mass [mg]'].transform('sum')

    # Compute surface per CPU from the first two valid size values
    df['Surface [mm²]'] = df.groupby('Name')['Size [mm]'].transform(compute_surface_from_first_two)

    return df

# Clean all DataFrames
dfs_high_level = [clean_bom_df(df) for df in dfs]

# Save cleaned DataFrames
output_path = r"C:\Users\Arthur Resilio\Desktop\dfs_cleaned.pkl"
pd.to_pickle(dfs_high_level, output_path)

```

## Helper function to calculate standard deviation from a list of percentages

```python
def calculate_stats(percentage_lists):
    """Calculate mean and standard deviation from a list of percentages."""
    stats = {}
    for key, percentages in percentage_lists.items():
        if percentages:
            stats[key] = {
                'Percentage': np.mean(percentages),
                'Std Dev': np.std(percentages)
            }
        else:
            stats[key] = {
                'Percentage': 0,
                'Std Dev': 0
            }
    return stats
```

## Helper function to calculate the covariance matrix between beta and gamma

Since they are both extracted from the same dataset there can be a correlation

```python
def calculate_covariance(dfs_clean):
    covariance_results = {}

    for socket in ['LGA', 'BGA']:
        # Combine all dataframes for the current socket
        socket_df = pd.concat([df[df['Socket'] == socket] for df in dfs_clean], ignore_index=True)
        if socket_df.empty:
            continue

        # Get unique high-level materials and substances
        high_level_materials = socket_df['High Level Material'].unique()
        substances = socket_df['Substances'].unique()

        # Initialize a dictionary to store the data for each high-level material and substance
        data_dict = {material: {sub: [] for sub in substances} for material in high_level_materials}

        # Populate the data dictionary with the percentages
        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    data_dict[material][sub].append(percentage)

        # Create a DataFrame for each sample (row in the original data)
        sample_data = []
        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            sample_data_row = {}
            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()
                sample_data_row[material] = total_mass / sample_socket_df['CPU Total Mass [mg]'].iloc[0] if sample_socket_df['CPU Total Mass [mg]'].iloc[0] else 0

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    sample_data_row[f"{material}_{sub}"] = percentage

            sample_data.append(sample_data_row)

        # Create a DataFrame from the sample data
        sample_df = pd.DataFrame(sample_data)

        # Calculate the covariance matrix
        cov_matrix = sample_df.cov().fillna(0)

        # Extract the covariance between beta (high-level materials) and gamma (substances)
        n_beta = len(high_level_materials)
        cov_beta = cov_matrix.loc[high_level_materials, high_level_materials].values
        cov_gamma = np.zeros((n_beta, n_beta))
        cov_beta_gamma = np.zeros((n_beta, n_beta))

        for i, material_i in enumerate(high_level_materials):
            for k, material_k in enumerate(high_level_materials):
                # Covariance between material_i and material_k's substances
                for sub in substances:
                    cov_gamma[i, k] += cov_matrix.loc[f"{material_i}_{sub}", f"{material_k}_{sub}"]
                    cov_beta_gamma[i, k] += cov_matrix.loc[material_i, f"{material_k}_{sub}"]

        # Normalize the covariance matrices
        cov_gamma /= len(substances)
        cov_beta_gamma /= len(substances)

        # Store the covariance matrices for the current socket
        covariance_results[socket] = {
            'cov_beta': cov_beta,
            'cov_gamma': cov_gamma,
            'cov_beta_gamma': cov_beta_gamma
        }

    return covariance_results
```

## **Calculation of $\vec\beta$**

$\vec\beta : (5 \times 1)$

$\Sigma_i \beta_i = 1$, i = 1,2,3,4,5 (die, connection, packaging, carrier, finishing)

Linear regression: % of the total mass of the CPU for each high level component


```python
def high_level_mass_matrix(dfs_clean):
    results = {}

    for socket in ['LGA', 'BGA']:
        percentage_lists = {material: [] for material in pd.concat(dfs_clean)['High Level Material'].unique()}

        for df in dfs_clean:
            socket_df = df[df['Socket'] == socket]
            if socket_df.empty:
                continue

            total_mass = socket_df['Mass [mg]'].sum()
            for material in percentage_lists.keys():
                material_mass = socket_df[socket_df['High Level Material'] == material]['Mass [mg]'].sum()
                percentage = material_mass / total_mass if total_mass else 0
                percentage_lists[material].append(percentage)

        stats = calculate_stats(percentage_lists)
        results[socket] = pd.DataFrame.from_dict(stats, orient='index')

    return results
```

## **Calculation of matrix **$\gamma$****

$\gamma : (5 \times A)$, lines are the 5 high level component, the columns are the A substances

$\Sigma_j \gamma_{ij} = 1$, for i each of the high level components i = 1,...,5 and j = 1,..., A ; with A the number total of substances

Linear regression: % of the mass of each substance inside each high level component

```python
def substance_mass_matrices(dfs_clean):
    results = {}

    for socket in ['LGA', 'BGA']:
        socket_df = pd.concat([df[df['Socket'] == socket] for df in dfs_clean], ignore_index=True)
        if socket_df.empty:
            continue

        high_level_materials = socket_df['High Level Material'].unique()
        substances = socket_df['Substances'].unique()

        percentage_lists = {material: {sub: [] for sub in substances} for material in high_level_materials}

        for df in dfs_clean:
            sample_socket_df = df[df['Socket'] == socket]
            if sample_socket_df.empty:
                continue

            for material in high_level_materials:
                material_df = sample_socket_df[sample_socket_df['High Level Material'] == material]
                total_mass = material_df['Mass [mg]'].sum()

                for sub in substances:
                    sub_mass = material_df[material_df['Substances'] == sub]['Mass [mg]'].sum()
                    percentage = sub_mass / total_mass if total_mass else 0
                    percentage_lists[material][sub].append(percentage)

        # Initialize matrices
        percentage_matrix = pd.DataFrame(index=high_level_materials, columns=substances)
        std_dev_matrix = pd.DataFrame(index=high_level_materials, columns=substances)

        for material in high_level_materials:
            for sub in substances:
                percentages = percentage_lists[material][sub]
                if percentages:
                    percentage_matrix.loc[material, sub] = np.mean(percentages)
                    std_dev_matrix.loc[material, sub] = np.std(percentages)
                else:
                    percentage_matrix.loc[material, sub] = 0
                    std_dev_matrix.loc[material, sub] = 0

        results[socket] = {
            'Percentage': percentage_matrix.fillna(0),
            'Std Dev': std_dev_matrix.fillna(0)
        }

    return results
```

## **Calculation of the model**

Inputs of the model (parameters) : 
- total CPU mass [mg] = $w$, (1 $\times$ 1)

Other inputs of the model and intermediate results :
- high level components mass [mg] = $ \vec c $, $c_i$ with i = 1, 2, ... 5 ; (1 $\times$ 5). Calculated using the total mass $w$ using the function $g$

Output of the model :
- substances mass [mg] = $\vec m  = \vec c \times \vec \gamma $ ; are the masses of each of the susbtances calculated using the function $h$

 (1 $\times$ 5) $\times$ (5 $\times$ A) = (1 $\times$ A) with A the number total of substances

*Formated to be use in the Resilio API with the corresponding Ecoinvent elementary flow*

### 1. Function g

Input :
- total CPU mass [mg] $w$ : (1 $\times$ 1)
- percentages of the total mass of the CPU for each high level component [%], $\vec \beta$ : (1 $\times$ 5),

Output :
- mass of each of the high level components [mg], $\vec c$ : (1 $\times$ 5)

$g(w, \vec \beta) = \vec c = \vec \beta \times \vec w  <=> c_i = \beta_i \times w_i $  with i = 1, 2, 3, 4, 5

```python
def g(w, beta):
    """Calculate the mass of each high-level component."""
    c = np.zeros(5)
    for i in range(0, 4):
        c[i] = beta[i] * w
    return c
```

### 2. Function h

Input :
- mass of each of the high level components [mg], $\vec c$ : (1 $\times$ 5)
- percentage of mass of each substance in each high level component [%], $\vec \gamma$ : (5 $\times$ A), A is the number of substances

Output :
- mass of each of the substances [mg], $\vec m$ : (1 $\times$ A), A is the number of substances

$h(\vec c, \vec \gamma) = \vec m = \vec c \times \vec \gamma$

```python
def h(c_vec, gamma_vec):
    """Calculate the mass of each of the substances."""
    return c_vec @ gamma_vec
```

### Modelisation

- $\vec c = g(w, \vec\beta) = \vec\beta \times w → c_i = \beta_i \times w$

- $m = h(\vec c, \gamma) = \vec c \times \gamma → m_j = \Sigma_i (c_i \times\gamma_{ij}) = \Sigma_i (\beta_i \times w \times \gamma_{ij})$

Definition of the mass of a substance j : $m_j = w \times \Sigma_i (\vec\beta_i \times γ_{ij})$

```python
def model(w, beta, gamma):
    high_levels = ['die', 'connection', 'packaging', 'carrier', 'finishing']
    beta = beta['Percentage'].values.flatten()
    c = np.array([b * w for b in beta])  # shape: [5]
    
    # Compute per-component substance masses: shape [5, n_substances]
    # Each row i = mass of each substance coming from high-level component i
    m_matrix = c[:, np.newaxis] * gamma.to_numpy()  # broadcast: [5,1] * [5, n_sub]
    
    # Build the DataFrame row by row across high-level components
    substance_masses_df = pd.DataFrame({
        'High Level': np.repeat(high_levels, len(gamma.columns)),
        'Substances': gamma.columns.tolist() * len(high_levels),
        'Mass [mg]': m_matrix.flatten()
    })
    return substance_masses_df
```

# **Error propagation in the model**

### **1. Model definition**

Steps of the model :

- $\vec c = g(w, \vec\beta) = \vec\beta \times w → c_i = \beta_i \times w$

- $m = h(\vec c, \gamma) = \vec c \times \gamma → m_j = \Sigma_i (c_i \times\gamma_{ij}) = \Sigma_i (\beta_i \times w \times \gamma_{ij})$

Definition of the mass of a substance j : $m_j = w \times \Sigma_i (\vec\beta_i \times γ_{ij})$

Incertainties :

- $\sigma_w$ (uncertainty of total mass w) ?
- $\sigma_{\beta_i}$ (uncertainty of each $\beta_i$)
- $\sigma_{\gamma_{ij}}$ (uncertainty of each $\gamma_{ij}$)

Assumptions : Independent variables (but $\gamma$ and $\beta$ are correlated since they both have been estimated from the same dataset)

### **2. Error propagation**

The general formula for variance propagation for a function $y = f(x_1, x_2, ...)$ is : 

$\sigma_y^2 = \Sigma_i (\frac{∂ f}{∂ x_i})^2 \times \sigma_{x_i}^2$

In our case : $m_j = f(\vec \beta, w, \gamma) = w \times \Sigma_i (\beta_i \times \gamma_{ij})$ thus,

$\sigma_{m_j}^2 = (\frac{∂m_j}{∂w})² σ_w² + Σᵢ (\frac{∂m_j}{∂βᵢ})² σ_{βᵢ}² + Σᵢ (\frac{∂m_j}{∂γᵢⱼ})² σ_{γᵢⱼ}² $

With :

- $∂m_j/∂w = Σᵢ (βᵢ × γᵢⱼ)$
- $∂m_j/∂βᵢ = w × γᵢⱼ$
- $∂m_j/∂γᵢⱼ = w × βᵢ$


**Law of variance propagation : $K_{yy} = F · K_{xx} · F^T$**

With:

- $K_{yy}$ = covariance matrix of outputs
- F = Jacobian matrix of partial derivatives
- $K_{xx}$ = covariance matrix of inputs



```python
def propagate_error(model_output_df, total_cpu_mass, sigma_w, beta_df, gamma_df, sigma_gamma_df, cov_beta_gamma=None):
    """
    Propagates uncertainty through the model:
        m_j = w * sum_i (beta_i * gamma_ij)

    Explicitly constructs the Jacobian matrix F and the covariance matrix K_xx,
    accounting for potential correlations between beta and gamma.

    Parameters:
    - model_output_df: DataFrame from model() with columns ['High Level', 'Substance', 'Mass [mg]']
    - total_cpu_mass: scalar, total CPU mass in mg
    - sigma_w: scalar, uncertainty of total_cpu_mass
    - beta_df: DataFrame with 'Percentage' and 'Std Dev' for high-level components
    - gamma_df: DataFrame with substance fractions (n_beta x A)
    - sigma_gamma_df: DataFrame with std dev of gamma (n_beta x A)
    - cov_beta_gamma: Covariance matrix between beta and gamma (optional, default is None)

    Returns:
    - model_output_df with added columns:
        'Uncertainty [mg]', 'Rel Uncertainty [%]'
    """

    # Extract beta and sigma_beta
    beta = beta_df['Percentage'].values  # shape (n_beta,)
    sigma_beta = beta_df['Std Dev'].values  # shape (n_beta,)

    # Extract gamma and sigma_gamma
    gamma = gamma_df.values  # shape (n_beta, A)
    sigma_gamma = sigma_gamma_df.values  # shape (n_beta, A)

    # Number of substances and high-level components
    A = gamma.shape[1]
    n_beta = beta.shape[0]

    # Get unique high-level materials (must match beta order)
    high_level_materials = beta_df.index.tolist() if hasattr(beta_df, 'index') else [f"HL{i}" for i in range(n_beta)]

    # Initialize uncertainty array — now we need one per (high-level, substance) pair
    # We'll compute it per high-level material and substance
    sigma_m = np.zeros(len(model_output_df))  # shape (500,)

    # For each row in model_output_df (each (high-level, substance) pair)
    for idx, row in model_output_df.iterrows():
        high_level = row['High Level']
        substance = row['Substances']

        # Get index of high-level material
        try:
            i = high_level_materials.index(high_level)
        except ValueError:
            # If high-level not in beta, skip or set 0
            sigma_m[idx] = 0.0
            continue

        # Get index of substance
        try:
            j = gamma_df.columns.get_loc(substance)
        except KeyError:
            sigma_m[idx] = 0.0
            continue

        # Construct Jacobian matrix F for this (i, j) pair
        dmdw = beta[i] * gamma[i, j]  # ∂m_ij/∂w
        dmdbeta = np.zeros(n_beta)
        dmdbeta[i] = total_cpu_mass * gamma[i, j]  # ∂m_ij/∂β_i
        dmdgamma = np.zeros(n_beta)
        dmdgamma[i] = total_cpu_mass * beta[i]  # ∂m_ij/∂γ_ij

        # Combine into F (1 x (1 + n_beta + n_beta) row vector)
        F = np.hstack([[dmdw], dmdbeta, dmdgamma])

        # Construct covariance matrix K_xx
        # Diagonal elements for w, beta, and gamma
        diag_elements = np.hstack([[sigma_w**2], sigma_beta**2, sigma_gamma[:, j]**2])

        # Initialize K_xx as a diagonal matrix
        K_xx = np.diag(diag_elements)

        # If cov_beta_gamma is provided, update the off-diagonal elements
        if cov_beta_gamma is not None:
            # Assume cov_beta_gamma is shape (n_beta, n_beta)
            for k in range(n_beta):
                for l in range(n_beta):
                    # Update covariance between beta_k and gamma_lj
                    K_xx[1 + k, 1 + n_beta + l] = cov_beta_gamma[k, l]
                    K_xx[1 + n_beta + l, 1 + k] = cov_beta_gamma[k, l]

        # Compute variance using F * K_xx * F^T
        sigma_m_sq = F @ K_xx @ F.T
        sigma_m[idx] = np.sqrt(sigma_m_sq)

    # Add to DataFrame
    result_df = model_output_df.copy()
    result_df['Uncertainty [mg]'] = sigma_m
    result_df['Rel Uncertainty [%]'] = (sigma_m / result_df['Mass [mg]']) * 100

    return result_df
```

### Since gamma and beta are estimated from the same set of material declaration sheets, we can take into account the correlation with covariance matrices, however due to small dataset it may exagerate the incertitudes

```python
def propagate_error_cov(model_output_df, total_cpu_mass, sigma_w, beta_df, gamma_df, sigma_gamma_df, cov_beta=None, cov_gamma=None, cov_beta_gamma=None):
    """
    Propagates uncertainty through the model using covariance matrices.

    Computes two bounds:
    - Underestimate: assumes no correlations (diagonal covariances)
    - Overestimate: uses provided covariances (if any)

    Parameters:
    - model_output_df: DataFrame with columns ['High Level', 'Substance', 'Mass [mg]']
    - total_cpu_mass: scalar, total CPU mass in mg
    - sigma_w: scalar, uncertainty of total_cpu_mass
    - beta_df: DataFrame with 'Percentage' and 'Std Dev' for high-level components
    - gamma_df: DataFrame with substance fractions (n_beta x A)
    - sigma_gamma_df: DataFrame with std dev of gamma (n_beta x A)
    - cov_beta: Covariance matrix for beta (n_beta x n_beta) (optional)
    - cov_gamma: Covariance matrix for gamma (n_beta x n_beta) (optional)
    - cov_beta_gamma: Covariance matrix between beta and gamma (optional, n_beta x n_beta)

    Returns:
    - model_output_df with added columns:
        'Sigma [mg] (Underestimate)', 'Sigma [%] (Underestimate)',
        'Sigma [mg] (Overestimate)', 'Sigma [%] (Overestimate)'
    """

    # Extract beta and sigma_beta
    beta = beta_df['Percentage'].values  # shape (n_beta,)
    sigma_beta = beta_df['Std Dev'].values  # shape (n_beta,)

    # Extract gamma and sigma_gamma
    gamma = gamma_df.values  # shape (n_beta, A)
    sigma_gamma = sigma_gamma_df.values  # shape (n_beta, A)

    # Number of substances and high-level components
    A = gamma.shape[1]
    n_beta = beta.shape[0]

    # Get high-level material names (must match beta order)
    high_level_materials = beta_df.index.tolist() if hasattr(beta_df, 'index') else [f"HL{i}" for i in range(n_beta)]

    # Initialize uncertainty arrays
    sigma_m_under = np.zeros(len(model_output_df))
    sigma_m_over = np.zeros(len(model_output_df))

    # For each row in model_output_df
    for idx, row in model_output_df.iterrows():
        high_level = row['High Level']
        substance = row['Substances']

        # Get index of high-level material
        try:
            i = high_level_materials.index(high_level)
        except ValueError:
            sigma_m_under[idx] = 0.0
            sigma_m_over[idx] = 0.0
            continue

        # Get index of substance
        try:
            j = gamma_df.columns.get_loc(substance)
        except KeyError:
            sigma_m_under[idx] = 0.0
            sigma_m_over[idx] = 0.0
            continue

        # Construct Jacobian matrix F for this (i, j) pair
        dmdw = beta[i] * gamma[i, j]  # ∂m_ij/∂w
        dmdbeta = np.zeros(n_beta)
        dmdbeta[i] = total_cpu_mass * gamma[i, j]  # ∂m_ij/∂β_i
        dmdgamma = np.zeros(n_beta)
        dmdgamma[i] = total_cpu_mass * beta[i]  # ∂m_ij/∂γ_ij

        # Combine into F (1 x (1 + n_beta + n_beta) row vector)
        F = np.hstack([[dmdw], dmdbeta, dmdgamma])

        # ————————————————————————————————————————————————————————————————
        # Underestimate: Assume no correlations (diagonal)
        # ————————————————————————————————————————————————————————————————

        K_w = np.array([[sigma_w**2]])
        K_beta_under = np.diag(sigma_beta**2)
        K_gamma_under = np.diag(sigma_gamma[:, j]**2)
        K_beta_gamma_under = np.zeros((n_beta, n_beta))

        K_xx_under = np.block([
            [K_w, np.zeros((1, n_beta)), np.zeros((1, n_beta))],
            [np.zeros((n_beta, 1)), K_beta_under, K_beta_gamma_under],
            [np.zeros((n_beta, 1)), K_beta_gamma_under.T, K_gamma_under]
        ])

        sigma_m_sq_under = F @ K_xx_under @ F.T
        sigma_m_under[idx] = np.sqrt(sigma_m_sq_under)

        # ————————————————————————————————————————————————————————————————
        # Overestimate: Use provided covariances (if any)
        # ————————————————————————————————————————————————————————————————

        # Use provided covariances if available, else fall back to diagonal
        K_beta_over = cov_beta if cov_beta is not None else np.diag(sigma_beta**2)
        K_gamma_over = cov_gamma if cov_gamma is not None else np.diag(sigma_gamma[:, j]**2)
        K_beta_gamma_over = cov_beta_gamma if cov_beta_gamma is not None else np.zeros((n_beta, n_beta))

        K_xx_over = np.block([
            [K_w, np.zeros((1, n_beta)), np.zeros((1, n_beta))],
            [np.zeros((n_beta, 1)), K_beta_over, K_beta_gamma_over],
            [np.zeros((n_beta, 1)), K_beta_gamma_over.T, K_gamma_over]
        ])

        sigma_m_sq_over = F @ K_xx_over @ F.T
        sigma_m_over[idx] = np.sqrt(sigma_m_sq_over)

    # Add to DataFrame
    result_df = model_output_df.copy()
    result_df['Sigma [mg] (Underestimate)'] = sigma_m_under
    result_df['Sigma [%] (Underestimate)'] = (sigma_m_under / result_df['Mass [mg]']) * 100
    result_df['Sigma [mg] (Overestimate)'] = sigma_m_over
    result_df['Sigma [%] (Overestimate)'] = (sigma_m_over / result_df['Mass [mg]']) * 100

    return result_df
```

# **Correction of the $\beta$ coefficients**

The goal is to update the packaging percentage using a weighted average of the old value and the new experience values. The weights are inversely proportional to the variances (squares of the standard deviations). We have 3 experiences thus $i=3$.

### **1. Weighted average to update the percentages**



From Section *4.4.1 (Cofacteurs des paramètres compensés)* and Section *4.8.4 (Poids et corrélations)*, we use the concept of weighted least squares to combine the old and new values: 

$\hat x = \frac{\Sigma^n_{i=1}w_i x_i}{\Sigma^n_{i=1}w_i}$
- $w_i = \frac{1}{\sigma^2_i}$ is the weight of each observation
- $x_i$ is the value of each observation
- $\sigma_i$ is the standard deviation of each observation

### **2. Propagation of uncertainty**

The standard deviation of the updated packaging percentage is calculated using the formula for the variance of a weighted average.

Formula from *Methodes d'estimation*:

From Section *4.4.1 (Cofacteurs des paramètres compensés)*:

$\sigma^2_{\hat x} = \frac{1}{\Sigma^n_{i=1}w_i}$ is the variance of the weighted average

### **3. Scaling percentages to ensure sum is 1**

After updating the packaging percentage, the other percentages are scaled so that the total sum of all percentages remains 1:

scale_factor = $\frac{1-\beta_{updated}}{\Sigma_{\beta_i}}$



```python
experiences = pd.DataFrame({
    'CPU mass (g)': [107.50, 105.90, 109.90],
    'Packaging mass (g)': [83.10, 83.00, 89.20],
    '%': [0.77, 0.78, 0.81]
})
```

```python
def update_packaging_beta(old_beta_df, experiences_df, precision_mg=0.1):
    """
    Update only the packaging beta percentage using new experience values and their precision.
    Scale the other percentages so that the total sum remains 1.

    Parameters:
    - old_beta_df: DataFrame with columns 'Percentage' and 'Std Dev' for old beta values.
    - experiences_df: DataFrame with columns 'CPU mass (g)', 'Packaging mass (g)', and '%'.
    - precision_mg: Precision of the new experience measurements in milligrams (default: 0.1 mg).

    Returns:
    - Updated beta DataFrame with new 'Percentage' and 'Std Dev'.
    """
    # Extract old beta values and their standard deviations
    old_percentages = old_beta_df['Percentage'].values
    old_std_devs = old_beta_df['Std Dev'].values

    # Extract new experience percentages and their standard deviations
    new_percentages = experiences_df['%'].values
    new_std_devs = np.full_like(new_percentages, precision_mg / 1000)  # Convert mg to g

    # Calculate weights for old and new packaging values
    old_packaging_weight = 1 / (old_std_devs[2] ** 2)  # Index 2 corresponds to 'packaging'
    new_packaging_weight = 1 / (new_std_devs ** 2)

    # Calculate the combined weight for packaging
    total_packaging_weight = old_packaging_weight + np.sum(new_packaging_weight)

    # Calculate the weighted average for packaging
    updated_packaging_percentage = (
        old_packaging_weight * old_percentages[2] +
        np.sum(new_packaging_weight * new_percentages)
    ) / total_packaging_weight

    # Calculate the updated standard deviation for packaging
    updated_packaging_std_dev = np.sqrt(1 / total_packaging_weight)

    # Create a copy of the old beta DataFrame
    updated_beta_df = old_beta_df.copy()

    # Update the packaging percentage and standard deviation
    updated_beta_df.at['packaging', 'Percentage'] = updated_packaging_percentage
    updated_beta_df.at['packaging', 'Std Dev'] = updated_packaging_std_dev

    # Calculate the sum of the other percentages
    other_percentages = updated_beta_df.loc[updated_beta_df.index != 'packaging', 'Percentage']
    sum_other_percentages = other_percentages.sum()

    # Scale the other percentages so that the total sum is 1
    scale_factor = (1 - updated_packaging_percentage) / sum_other_percentages
    updated_beta_df.loc[updated_beta_df.index != 'packaging', 'Percentage'] *= scale_factor

    return updated_beta_df
```

```python
def calculate_proportions(df):
    """
    Calculate the proportion of each substance relative to the sample weight,
    and the standard deviation of those proportions.

    Parameters:
    - df: pandas DataFrame with columns for sample weight, substances, and their standard deviations.

    Returns:
    - sample_gamma: DataFrame with proportions.
    - sample_gamma_variance: DataFrame with standard deviations of proportions.
    """
    # Convert all columns to numeric
    df = df.apply(pd.to_numeric)

    # Calculate proportions
    sample_gamma = df.drop(columns=[col for col in df.columns if col.startswith('Std_')]).copy()
    for col in sample_gamma.columns:
        if col != "Sample weight (mg)":
            sample_gamma[col] = sample_gamma[col] / sample_gamma["Sample weight (mg)"]

    sample_gamma = sample_gamma.drop(columns=["Sample weight (mg)", "Error"])
    sample_gamma = rename_columns(sample_gamma)

    return sample_gamma
```

```python
def calculate_relative_uncertainty(df):
    # Convert relevant columns to numeric
    numeric_cols = ['Sample weight (mg)', 'Error'] + [col for col in df.columns if col not in ['Sample weight (mg)', 'Error']]
    df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

    # Create a new DataFrame to store the results
    result_df = pd.DataFrame()
    result_df = result_df.apply(pd.to_numeric, errors='coerce')

    # Iterate over each sample (row)
    for index, row in df.iterrows():
        sample_weight = row['Sample weight (mg)']
        error = row['Error']

        # Iterate over each substance and its standard deviation
        substances = []
        for col in df.columns[2::2]:  # Skip the first two columns (Sample weight and Error)
            substance = col
            std_col = f"Std_{substance}"

            substance_value = row[col]
            std_value = row[std_col]

            # Calculate the relative uncertainty
            relative_uncertainty = substance_value * np.sqrt(
                (error / sample_weight)**2 + (std_value / substance_value)**2
            )

            substances.append(relative_uncertainty)

        # Add the calculated uncertainties to the result DataFrame
        result_df.loc[index, 'Sample weight (mg)'] = sample_weight
        result_df.loc[index, 'Error'] = error
        for i, col in enumerate(df.columns[2::2]):
            result_df.loc[index, col] = substances[i]

    result_df = result_df.drop(columns=["Sample weight (mg)", "Error"])
    result_df = result_df.fillna(result_df.max().max())
    result_df = rename_columns(result_df)       

    return result_df
```

## **$\gamma$ update**

This function implements parametric least squares adjustment as taught in Chapter 4 of : *Compensation paramétrique*

The function *update_gamma_parametric_adjustment*:
- Treats prior values as pseudo-observations;
- Treats experimental data as real observations;
- Uses weighted least squares to compute optimal parameter estimates;
- Computes standard deviations using a posteriori variance;
- Is linear, so no iteration needed;
- Is performed per substance, independently.

Input :
- $\gamma$ the coefficients to correct $(4 \times 15)$ for the 4 categories and 15 substances 
- $\gamma _{std}$ the standard error of the coefficients $(4 \times 15)$ 
- $experience$ the experience proportion for each material for each category $(4 \times 15)$
- $experience_{std}$ the experience stand error for each proportion $(4 \times 15)$

Output:
- $\gamma_{updated}$ updated coefficients $(4 \times 15)$ for the 4 categories and 15 substances 
- $\gamma _{std}{updated}$ the standard error of the updated coefficients $(4 \times 15)$ 


**(Ch. 4: Compensation paramétrique)**

We estimate parameters $\mathbf{x}$ from observations $\boldsymbol{\ell}$ using:

$$
\boldsymbol{\ell} - \mathbf{v} = \mathbf{f}(\mathbf{x})
$$

For **direct observations**:  
$$
\ell_i - v_i = x_j \quad \Rightarrow \quad \mathbf{f}(\mathbf{x}) = x_j
$$


### Key formulas

**Design matrix** $$\mathbf{A}$$ (size $n \times u$): Each row has 1 at column $j$ (part index), 0 elsewhere.

**Approximate residuals**:  
$$
\mathbf{v}^{\circ} = \boldsymbol{\ell} - \mathbf{x}^{\circ}[j]
$$

**Weight matrix**:  
$$
\mathbf{P} = \text{diag}(1 / \sigma_i^2)
$$

**Parameter increments** (Eq. 4.13):  
$$
\Delta \hat{\mathbf{x}} = \left( \mathbf{A}^\top \mathbf{P} \mathbf{A} \right)^{-1} \mathbf{A}^\top \mathbf{P} \mathbf{v}^{\circ}
$$

**Compensated parameters**:  
$$
\hat{\mathbf{x}} = \mathbf{x}^{\circ} + \Delta \hat{\mathbf{x}}
$$

**Cofactor matrix** (Eq. 4.20):  
$$
\mathbf{Q}_{\hat{\mathbf{x}}\hat{\mathbf{x}}} = \left( \mathbf{A}^\top \mathbf{P} \mathbf{A} \right)^{-1}
$$

**A posteriori variance** (Eq. 4.27):  
$$
\hat{\sigma}_0^2 = \frac{\hat{\mathbf{v}}^\top \mathbf{P} \hat{\mathbf{v}}}{n - u}, \quad \hat{\mathbf{v}} = \mathbf{v}^{\circ} - \mathbf{A} \Delta \hat{\mathbf{x}}
$$

**Covariance matrix**:  
$$
\mathbf{K}_{\hat{\mathbf{x}}\hat{\mathbf{x}}} = \hat{\sigma}_0^2 \cdot \mathbf{Q}_{\hat{\mathbf{x}}\hat{\mathbf{x}}}
$$

**Standard deviations**:  
$$
\sigma_{\hat{x}_j} = \sqrt{ \left( \mathbf{K}_{\hat{\mathbf{x}}\hat{\mathbf{x}}} \right)_{jj} }
$$

---

*Assumptions: Linear model, independent observations, $n > u$ (redundancy), weights = $\sigma_i^2$.*

*All formulas from Chapter 4 of course material.*

This function combines prior estimates and experimental measurements into statistically optimal parameter updates using weighted least squares, treating both as observations with known uncertainties. 

It assigns higher weight to more precise data (smaller σ), letting them influence the result more.  

The math finds the parameter values that minimize the weighted sum of squared residuals, following Chapter 4’s parametric adjustment.  

It then updates both the parameter values and their uncertainties, reducing error through redundancy.  


```python
import numpy as np
import pandas as pd
from typing import List, Tuple, Union


def update_gamma_parametric_adjustment(
    gamma: pd.DataFrame,
    gamma_std: pd.DataFrame,
    experience: List[pd.DataFrame],
    experience_std: List[pd.DataFrame]
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Update gamma coefficients using parametric least squares adjustment.
    Combines prior gamma values with experimental data as weighted observations.
    
    Based on Chapter 4: Compensation paramétrique
    
    Model: ℓ - v = f(x)
    For direct observations: ℓ - v = x
    
    Parameters:
    -----------
    gamma : pd.DataFrame
        Prior gamma values (4 parts × 15 substances)
        Used as both approximate parameters x̊ and pseudo-observations
    gamma_std : pd.DataFrame
        Standard deviations of prior gamma values (4 × 15)
    experience : List[pd.DataFrame]
        Experimental measurements, each (4 × 15)
    experience_std : List[pd.DataFrame]
        Standard deviations of experimental measurements
    
    Returns:
    --------
    Tuple[pd.DataFrame, pd.DataFrame]
        (x̂, σ̂_x): Compensated parameters and their standard deviations
    
    References:
    -----------
    - Equation 4.13: x̂ = (AᵀPA)⁻¹ AᵀPv̊
    - Equation 4.20: Q_x̂x̂ = (AᵀPA)⁻¹
    - Equation 4.27: σ̂₀² = v̂ᵀPv̂ / (n-u)
    """
    
    n_parts = gamma.shape[0]        # u = 4 parameters (parts)
    n_substances = gamma.shape[1]   # 15 substances
    n_samples = len(experience)     # number of experimental samples
    
    # Initialize output dataframes
    x_hat = gamma.copy()  # Compensated parameters
    sigma_x_hat = gamma_std.copy()  # Standard deviations of compensated parameters
    
    # Process each substance independently
    for j in range(n_substances):
        substance_name = gamma.columns[j]
        
        # ============================================================
        # Step 1: Define approximate parameters x̊
        # ============================================================
        x_approx = gamma.iloc[:, j].values  # x̊ (4,) - one value per part
        u = len(x_approx)  # u = 4 parameters
        
        # ============================================================
        # Step 2: Build observation vector ℓ and weight matrix P
        # ============================================================
        ell = []  # ℓ: observations
        sigma_squared = []  # σ²ᵢ: variances
        part_indices = []  # Track which part each observation corresponds to
        
        # Add prior as pseudo-observations
        ell_prior = gamma.iloc[:, j].values
        sigma_prior = gamma_std.iloc[:, j].values
        
        for part_idx in range(n_parts):
            if not np.isnan(ell_prior[part_idx]) and not np.isnan(sigma_prior[part_idx]):
                ell.append(ell_prior[part_idx])
                sigma_squared.append(sigma_prior[part_idx]**2)
                part_indices.append(part_idx)
        
        # Add experimental observations
        for k in range(n_samples):
            ell_exp = experience[k].iloc[:, j].values
            sigma_exp = experience_std[k].iloc[:, j].values
            
            for part_idx in range(n_parts):
                if not np.isnan(ell_exp[part_idx]) and not np.isnan(sigma_exp[part_idx]):
                    ell.append(ell_exp[part_idx])
                    sigma_squared.append(sigma_exp[part_idx]**2)
                    part_indices.append(part_idx)
        
        ell = np.array(ell)  # (n,)
        part_indices = np.array(part_indices)  # (n,)
        n = len(ell)  # n: number of observations
        
        if n <= u:
            print(f"Warning: Not enough observations for substance '{substance_name}' "
                  f"(n={n}, u={u}). Keeping original values.")
            continue
        
        # Weight matrix P = diag(1/σ²ᵢ)
        # Add small epsilon to avoid division by zero
        epsilon = 1e-10
        p_diag = 1.0 / (np.array(sigma_squared) + epsilon)
        P = np.diag(p_diag)  # (n × n)
        
        # ============================================================
        # Step 3: Build design matrix A
        # ============================================================
        # For direct observations: ℓᵢ - vᵢ = xⱼ
        # where j is the part index (0, 1, 2, or 3)
        # A is (n × u) matrix
        A = np.zeros((n, u))
        for i in range(n):
            part_idx = part_indices[i]
            A[i, part_idx] = 1.0
        
        # ============================================================
        # Step 4: Compute approximate residuals v̊
        # ============================================================
        # v̊ = ℓ - f(x̊)
        # For direct observations: f(x̊) = x̊[part_index]
        f_x_approx = x_approx[part_indices]  # (n,) - select appropriate x̊ for each obs
        v_approx = ell - f_x_approx  # v̊ (n,)
        
        # ============================================================
        # Step 5: Solve normal equations (Equation 4.13 and 4.20)
        # ============================================================
        # Normal matrix: N = AᵀPA
        N = A.T @ P @ A  # (u × u)
        
        # Right-hand side: b = AᵀPv̊
        b = A.T @ P @ v_approx  # (u,)
        
        try:
            # Cofactor matrix of parameters (Equation 4.20)
            Q_x_hat_x_hat = np.linalg.inv(N)  # (u × u)
            
            # Parameter increments (Equation 4.13)
            delta_x = Q_x_hat_x_hat @ b  # Δx̂ (u,)
            
            # Compensated parameters (Equation 4.14)
            x_compensated = x_approx + delta_x  # x̂ = x̊ + Δx̂ (u,)
            
            # ============================================================
            # Step 6: Compute compensated residuals and variance factor
            # ============================================================
            # Compensated residuals (Equation 4.16)
            v_hat = v_approx - A @ delta_x  # v̂ = v̊ - A·Δx̂ (n,)
            
            # Redundancy (surdétermination)
            r = n - u  # r = n - u
            
            # A posteriori variance factor (Equation 4.27)
            if r > 0:
                omega_hat = v_hat.T @ P @ v_hat  # Ω̂ = v̂ᵀPv̂
                sigma_0_squared_hat = omega_hat / r  # σ̂₀² = Ω̂ / r
            else:
                # Not enough redundancy - use a priori variance
                sigma_0_squared_hat = 1.0
                print(f"Warning: No redundancy for substance '{substance_name}'. "
                      f"Using a priori variance.")
            
            # ============================================================
            # Step 7: Compute covariance matrix of parameters
            # ============================================================
            # Covariance matrix: K_x̂x̂ = σ̂₀² · Q_x̂x̂
            K_x_hat_x_hat = sigma_0_squared_hat * Q_x_hat_x_hat  # (u × u)
            
            # Standard deviations of compensated parameters
            sigma_x_compensated = np.sqrt(np.diag(K_x_hat_x_hat))  # (u,)
            
            # ============================================================
            # Step 8: Store results
            # ============================================================
            x_hat.iloc[:, j] = x_compensated
            sigma_x_hat.iloc[:, j] = sigma_x_compensated
            
        except np.linalg.LinAlgError:
            # Singular matrix - keep original values
            print(f"Warning: Singular normal matrix for substance '{substance_name}'. "
                  f"Keeping original values.")
            continue
    
    return x_hat, sigma_x_hat
```

```python
def update_gamma_single_category(
    gamma: Union[pd.Series, pd.DataFrame],
    gamma_std: Union[pd.Series, pd.DataFrame],
    experience: Union[pd.DataFrame, List[Union[pd.Series, pd.DataFrame]]],
    experience_std: Union[pd.DataFrame, List[Union[pd.Series, pd.DataFrame]]]
) -> Tuple[Union[pd.Series, pd.DataFrame], Union[pd.Series, pd.DataFrame]]:
    """
    Update gamma coefficients using parametric least squares adjustment.
    Single category version (no parts subdivision).
    
    Based on Chapter 4: Compensation paramétrique
    
    Model: ℓ - v = f(x)
    For direct observations: ℓ - v = x
    
    Parameters:
    -----------
    gamma : pd.Series or pd.DataFrame
        Prior gamma values (15 substances)
        If DataFrame: assumes 1 row × 15 columns
        If Series: 15 elements
    gamma_std : pd.Series or pd.DataFrame
        Standard deviations of prior gamma values
    experience : pd.DataFrame or List[pd.Series/pd.DataFrame]
        Experimental measurements
        If DataFrame: each row is a sample, columns are substances
        If List: each element is a sample (Series or DataFrame with 1 row)
    experience_std : pd.DataFrame or List[pd.Series/pd.DataFrame]
        Standard deviations of experimental measurements
    
    Returns:
    --------
    Tuple[pd.Series or pd.DataFrame, pd.Series or pd.DataFrame]
        (x̂, σ̂_x): Compensated parameters and their standard deviations
    
    References:
    -----------
    - Equation 4.13: x̂ = (AᵀPA)⁻¹ AᵀPv̊
    - Equation 4.20: Q_x̂x̂ = (AᵀPA)⁻¹
    - Equation 4.27: σ̂₀² = v̂ᵀPv̂ / (n-u)
    
    Example:
    --------
    For a new 5th category with 2 experimental samples:
    - gamma: (1 × 15) prior values for the new category
    - experience: (2 × 15) two samples, 15 substances each
    - Total observations per substance: 1 (prior) + 2 (samples) = 3
    - Redundancy per substance: 3 - 1 = 2
    """
    
    # ============================================================
    # Input validation and format conversion
    # ============================================================
    
    # Convert gamma to Series if DataFrame
    is_dataframe = isinstance(gamma, pd.DataFrame)
    
    if is_dataframe:
        gamma_series = gamma.iloc[0, :]
        gamma_std_series = gamma_std.iloc[0, :]
    else:
        gamma_series = gamma
        gamma_std_series = gamma_std
    
    # Convert experience to list of Series
    experience_series = []
    experience_std_series = []
    
    # Case 1: experience is a DataFrame (each row is a sample)
    if isinstance(experience, pd.DataFrame):
        n_samples = len(experience)
        for i in range(n_samples):
            experience_series.append(experience.iloc[i, :])
            experience_std_series.append(experience_std.iloc[i, :])
    
    # Case 2: experience is already a list
    elif isinstance(experience, list):
        for k, exp in enumerate(experience):
            if isinstance(exp, pd.DataFrame):
                experience_series.append(exp.iloc[0, :])
            elif isinstance(exp, pd.Series):
                experience_series.append(exp)
            else:
                raise TypeError(f"experience[{k}] has unexpected type: {type(exp)}")
        
        for k, exp_std in enumerate(experience_std):
            if isinstance(exp_std, pd.DataFrame):
                experience_std_series.append(exp_std.iloc[0, :])
            elif isinstance(exp_std, pd.Series):
                experience_std_series.append(exp_std)
            else:
                raise TypeError(f"experience_std[{k}] has unexpected type: {type(exp_std)}")
    else:
        raise TypeError(f"experience has unexpected type: {type(experience)}")
    
    n_substances = len(gamma_series)  # 15 substances
    n_samples = len(experience_series)  # number of experimental samples
    
    # Initialize output Series
    x_hat = gamma_series.copy()  # Compensated parameters
    sigma_x_hat = gamma_std_series.copy()  # Standard deviations of compensated parameters
    
    # Process each substance independently
    for j in range(n_substances):
        substance_name = gamma_series.index[j]
        
        # ============================================================
        # Step 1: Define approximate parameter x̊
        # ============================================================
        x_approx = gamma_series.iloc[j]  # x̊ - scalar value for this substance
        u = 1  # u = 1 parameter per substance
        
        # ============================================================
        # Step 2: Build observation vector ℓ and weight matrix P
        # ============================================================
        ell = []  # ℓ: observations
        sigma_squared = []  # σ²ᵢ: variances
        
        # Add prior as pseudo-observation
        ell_prior = gamma_series.iloc[j]
        sigma_prior = gamma_std_series.iloc[j]
        
        # Check if values are valid (not NaN and not inf)
        if pd.notna(ell_prior) and pd.notna(sigma_prior) and np.isfinite(ell_prior) and np.isfinite(sigma_prior):
            ell.append(float(ell_prior))
            sigma_squared.append(float(sigma_prior)**2)
        
        # Add experimental observations
        for k in range(n_samples):
            ell_exp = experience_series[k].iloc[j]
            sigma_exp = experience_std_series[k].iloc[j]
            
            if pd.notna(ell_exp) and pd.notna(sigma_exp) and np.isfinite(ell_exp) and np.isfinite(sigma_exp):
                ell.append(float(ell_exp))
                sigma_squared.append(float(sigma_exp)**2)
        
        ell = np.array(ell)  # (n,)
        n = len(ell)  # n: number of observations
        
        if n <= u:
            print(f"Warning: Not enough observations for substance '{substance_name}' "
                  f"(n={n}, u={u}). Keeping original value.")
            continue
        
        # Weight matrix P = diag(1/σ²ᵢ)
        epsilon = 1e-10
        p_diag = 1.0 / (np.array(sigma_squared) + epsilon)
        P = np.diag(p_diag)  # (n × n)
        
        # ============================================================
        # Step 3: Build design matrix A
        # ============================================================
        # For direct observations of a single parameter: ℓᵢ - vᵢ = x
        # A is (n × 1) column vector of ones
        A = np.ones((n, 1))  # (n × 1)
        
        # ============================================================
        # Step 4: Compute approximate residuals v̊
        # ============================================================
        # v̊ = ℓ - f(x̊)
        # For direct observations: f(x̊) = x̊
        f_x_approx = np.full(n, x_approx)  # (n,) - all observations measure same parameter
        v_approx = ell - f_x_approx  # v̊ (n,)
        
        # ============================================================
        # Step 5: Solve normal equations (Equation 4.13 and 4.20)
        # ============================================================
        # Normal matrix: N = AᵀPA (scalar for u=1)
        N = A.T @ P @ A  # (1 × 1) - scalar
        N_scalar = N[0, 0]
        
        # Right-hand side: b = AᵀPv̊
        b = A.T @ P @ v_approx  # (1,)
        b_scalar = b[0]
        
        try:
            # Cofactor (scalar for u=1) (Equation 4.20)
            Q_x_hat_x_hat = 1.0 / N_scalar  # scalar
            
            # Parameter increment (Equation 4.13)
            delta_x = Q_x_hat_x_hat * b_scalar  # scalar
            
            # Compensated parameter (Equation 4.14)
            x_compensated = x_approx + delta_x  # x̂ = x̊ + Δx̂
            
            # ============================================================
            # Step 6: Compute compensated residuals and variance factor
            # ============================================================
            # Compensated residuals (Equation 4.16)
            v_hat = v_approx - A.flatten() * delta_x  # v̂ = v̊ - A·Δx̂ (n,)
            
            # Redundancy (surdétermination)
            r = n - u  # r = n - 1
            
            # A posteriori variance factor (Equation 4.27)
            if r > 0:
                omega_hat = v_hat.T @ P @ v_hat  # Ω̂ = v̂ᵀPv̂
                sigma_0_squared_hat = omega_hat / r  # σ̂₀² = Ω̂ / r
            else:
                sigma_0_squared_hat = 1.0
            
            # ============================================================
            # Step 7: Compute variance of parameter
            # ============================================================
            # Variance: σ²_x̂ = σ̂₀² · Q_x̂x̂
            variance_x_hat = sigma_0_squared_hat * Q_x_hat_x_hat  # scalar
            
            # Standard deviation of compensated parameter
            sigma_x_compensated = np.sqrt(variance_x_hat)
            
            # ============================================================
            # Step 8: Store results
            # ============================================================
            x_hat.iloc[j] = x_compensated
            sigma_x_hat.iloc[j] = sigma_x_compensated
            
        except (ZeroDivisionError, np.linalg.LinAlgError) as e:
            print(f"Warning: Cannot invert normal matrix for substance '{substance_name}': {e}. "
                  f"Keeping original value.")
            continue
    
    # Convert back to DataFrame if input was DataFrame
    if is_dataframe:
        x_hat_df = pd.DataFrame([x_hat.values], 
                                columns=gamma.columns, 
                                index=gamma.index)
        sigma_x_hat_df = pd.DataFrame([sigma_x_hat.values], 
                                      columns=gamma_std.columns, 
                                      index=gamma_std.index)
        return x_hat_df, sigma_x_hat_df
    else:
        return x_hat, sigma_x_hat
```
