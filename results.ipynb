```python
import scipy.stats as stats

df = pd.DataFrame(data)

# Define groups
gold = df[df["Group"] == "Gold"]
silver = df[df["Group"] == "Silver"]
amd = df[df["Group"] == "AMD"]
metal_lid = df[df["Group"] == "Metal Lid"]

# Function for t-tests
def ttest_summary(group1, group2, group1_name, group2_name):
    results = {}
    for element in ["Al", "Cu", "Fe", "Ni", "Zn", "Au", "Ti", "Sn"]:
        t, p = stats.ttest_ind(group1[element], group2[element], nan_policy='omit')
        results[element] = {"t": t, "p": p}
    return pd.DataFrame(results).T.rename(columns={"t": "t-statistic", "p": "p-value"})

# Perform t-tests
print("Gold vs. Silver:")
print(ttest_summary(gold, silver, "Gold", "Silver"))
print("\nGold vs. AMD:")
print(ttest_summary(gold, amd, "Gold", "AMD"))
print("\nSilver vs. AMD:")
print(ttest_summary(silver, amd, "Silver", "AMD"))
print("\nCPUs vs. Metal Lids:")
print(ttest_summary(pd.concat([gold, silver, amd, df[df["Group"] == "Bronze"]]), metal_lid, "CPUs", "Metal Lids"))
```
